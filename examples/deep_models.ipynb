# matchify - deep models

BERT and Siamese (fine-tuned BERT, contrastive loss) on the DBLP-ACM benchmark.

Requires `pip install matchify[deep]`.

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from matchify.models.bert_match_model import BertMatchModel
from matchify.models.siamese_match_model import SiameseMatchModel

df = pd.read_csv('../datasets/DBLP-ACM/CombinedAcademic.csv').head(800)
ignored = ['id', 'group_id', 'original_id']

field_config = {
    'title':   {'type': 'other'},
    'authors': {'type': 'other'},
    'venue':   {'type': 'other'},
    'year':    {'type': 'other'},
}
blocking_config = {'title': {'method': 'prefix', 'threshold': 3}}
```

```python
bert = BertMatchModel(df.copy(), field_config=field_config, blocking_config=blocking_config, ignored_columns=ignored)
bert.train()  # encode all records once
print(f'BertMatchModel MRR: {bert.mrr():.4f}')
```

```python
siamese = SiameseMatchModel(
    df.copy(),
    field_config=field_config,
    blocking_config=blocking_config,
    ignored_columns=ignored,
    n_pairs=2000,
    epochs=1,
)
siamese.train()  # fine-tune the encoder, then cache embeddings
print(f'SiameseMatchModel MRR: {siamese.mrr():.4f}')
```
