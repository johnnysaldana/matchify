# matchify quickstart

Train and evaluate every bundled model on the Amazon-Google products benchmark.

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from matchify.models.exact_match_model import ExactMatchModel
from matchify.models.flex_match_model import FlexMatchModel
from matchify.models.mlp_match_model import MLPMatchModel

df = pd.read_csv('../datasets/Amazon-GoogleProducts/CombinedProducts.csv').head(800)
ignored = ['id', 'group_id', 'original_id']

field_config = {
    'name':         {'type': 'other', 'comparison_method': 'jaro_winkler'},
    'description':  {'type': 'other', 'comparison_method': 'tfidf_cosine'},
    'manufacturer': {'type': 'other', 'comparison_method': 'jaro_winkler'},
    'price':        {'type': 'other', 'comparison_method': 'jaro_winkler'},
}
blocking_config = {'name': {'method': 'prefix', 'threshold': 2}}
```

```python
EMM = ExactMatchModel(df.copy(), ignored_columns=ignored)
print(f'ExactMatchModel MRR: {EMM.mrr():.4f}')
```

```python
FMM = FlexMatchModel(df.copy(), field_config=field_config, blocking_config=blocking_config, ignored_columns=ignored)
FMM.train()
print(f'FlexMatchModel MRR: {FMM.mrr():.4f}')
```

```python
MLP = MLPMatchModel(df.copy(), field_config=field_config, blocking_config=blocking_config, ignored_columns=ignored, n_pairs=2000)
MLP.train()
print(f'MLPMatchModel MRR: {MLP.mrr():.4f}')
```

```python
# Look up a single record
record = df[df['id'] == 8].iloc[0].drop(labels=ignored)
FMM.predict(record, only_matches=False, return_full_record=True).head(10)
```
